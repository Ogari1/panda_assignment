```python
# 1 Import the necessary libraries and load the dataset.
import pandas as pd

df = pd.read_csv("vgsales.csv")
```

```python
#Display the first 10 rows
df.head(10)
```

```python
#Dataframe shape 
shape = df.shape
print("shape:", shape)

#Dataframe basic formation
bf = df.info()
print("Basic formation:", bf)
```

```python
#Detect missing data

df.isna()
```

```python
df_drop = df.dropna()
print(df_drop)

df_fill = df.fillna(0)
print(df_fill)
```

```python
#Top5 video games
top5 = df.nlargest(5, 'Global_Sales')
print(top5)
```

```python
#Total sales in North America for the 'Action' genre

NA_action_sales = df[df['Genre'] == 'Action']['NA_Sales'].sum()
print(NA_action_sales)
```

```python
#top 3 publishers by global sales
top3_publishers = df.groupby('Publisher')['Global_Sales'].sum().nlargest(3)
print(top3_publishers)
```

```python
#average sales in Europe for the 'Shooter' genre

avg_shooter_sales_eu = df[df['Genre'] == 'Shooter']['EU_Sales'].mean()
print(avg_shooter_sales_eu)
```

```python
#game with the highest sales in Japan in the 'Sports' genre

highest_sales_jp_sports = df[df['Genre'] == 'Sports'].nlargest(1, 'JP_Sales')
print(highest_sales_jp_sports)
```
